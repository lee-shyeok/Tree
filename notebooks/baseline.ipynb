```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv("../data/train.csv")

# ================================
# 1. 기본 정보
# ================================
print("=== 데이터 크기 ===")
print(df.shape)

print("\n=== 결측치 현황 (상위 20개) ===")
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'결측치 수': missing, '결측치 비율(%)': missing_pct})
print(missing_df[missing_df['결측치 수'] > 0].sort_values('결측치 비율(%)', ascending=False).head(20))

print("\n=== 타겟 분포 ===")
print(df['임신 성공 여부'].value_counts())
print(df['임신 성공 여부'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')

# ================================
# 2. 타겟 분포 시각화
# ================================
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 타겟 비율
df['임신 성공 여부'].value_counts().plot(
    kind='bar', ax=axes[0], color=['#FF6B6B', '#4ECDC4']
)
axes[0].set_title('임신 성공 여부 분포')
axes[0].set_xlabel('0: 실패 / 1: 성공')
axes[0].set_ylabel('건수')

# 파이차트
df['임신 성공 여부'].value_counts().plot(
    kind='pie', ax=axes[1], autopct='%1.1f%%',
    labels=['실패', '성공'], colors=['#FF6B6B', '#4ECDC4']
)
axes[1].set_title('임신 성공 비율')

plt.tight_layout()
plt.savefig('../data/eda_target.png', dpi=100, bbox_inches='tight')
plt.show()

# ================================
# 3. 나이대별 임신 성공률
# ================================
age_success = df.groupby('시술 당시 나이')['임신 성공 여부'].mean().mul(100).round(2)
print("\n=== 나이대별 임신 성공률(%) ===")
print(age_success.sort_values(ascending=False))

plt.figure(figsize=(10, 5))
age_success.sort_index().plot(kind='bar', color='#4ECDC4')
plt.title('나이대별 임신 성공률')
plt.xlabel('나이대')
plt.ylabel('성공률 (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../data/eda_age.png', dpi=100, bbox_inches='tight')
plt.show()

# ================================
# 4. 시술 유형별 임신 성공률
# ================================
type_success = df.groupby('시술 유형')['임신 성공 여부'].mean().mul(100).round(2)
print("\n=== 시술 유형별 임신 성공률(%) ===")
print(type_success.sort_values(ascending=False))

plt.figure(figsize=(8, 4))
type_success.sort_values().plot(kind='barh', color='#45B7D1')
plt.title('시술 유형별 임신 성공률')
plt.xlabel('성공률 (%)')
plt.tight_layout()
plt.savefig('../data/eda_type.png', dpi=100, bbox_inches='tight')
plt.show()

# ================================
# 5. 수치형 컬럼 분포
# ================================
num_cols = ['총 시술 횟수', 'IVF 시술 횟수', '총 임신 횟수', '이식된 배아 수', 
            '총 생성 배아 수', '수집된 신선 난자 수']

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    df[col].hist(ax=axes[i], bins=30, color='#96CEB4', edgecolor='white')
    axes[i].set_title(col)
    axes[i].set_ylabel('건수')

plt.suptitle('주요 수치형 변수 분포', fontsize=14)
plt.tight_layout()
plt.savefig('../data/eda_numeric.png', dpi=100, bbox_inches='tight')
plt.show()
```

```python

# 결측치 처리


# 1. 99% 이상 결측 컬럼 삭제
drop_cols = [
    '난자 해동 경과일',
    'PGS 시술 여부',
    'PGD 시술 여부',
    '착상 전 유전 검사 사용 여부',
    '임신 시도 또는 마지막 임신 경과 연수'
]

df_clean = df.drop(columns=drop_cols)
print(f"삭제 후 컬럼 수: {df_clean.shape[1]}개 ({df.shape[1]} → {df_clean.shape[1]})")


# 2. 나머지 결측치 처리


# 수치형 컬럼 → 중앙값으로 채우기
num_cols = df_clean.select_dtypes(include='number').columns.tolist()
num_cols.remove('임신 성공 여부')  # 타겟은 제외

for col in num_cols:
    if df_clean[col].isnull().sum() > 0:
        median_val = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_val)
        print(f"[수치형] {col} → 중앙값 {median_val} 으로 채움")

# 범주형 컬럼 → 최빈값으로 채우기
cat_cols = df_clean.select_dtypes(include='object').columns.tolist()

for col in cat_cols:
    if df_clean[col].isnull().sum() > 0:
        mode_val = df_clean[col].mode()[0]
        df_clean[col] = df_clean[col].fillna(mode_val)
        print(f"[범주형] {col} → 최빈값 '{mode_val}' 으로 채움")


# 3. 처리 결과 확인

print("\n=== 결측치 처리 후 확인 ===")
remaining = df_clean.isnull().sum().sum()
print(f"남은 결측치 수: {remaining}개")

if remaining == 0:
    print("결측치 처리 완료!")
else:
    print("아직 결측치 남아있음:")
    print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])


# 4. 처리된 데이터 저장

df_clean.to_csv("../data/train_clean.csv", index=False, encoding='utf-8-sig')
print("\n train_clean.csv 저장 완료!")
print(f"최종 데이터 크기: {df_clean.shape}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df_clean = pd.read_csv("../data/train_clean.csv")


# 1. 수치형 변수만 추출

num_cols = df_clean.select_dtypes(include='number').columns.tolist()
print(f"수치형 변수 수: {len(num_cols)}개")
print(num_cols)


# 2. 타겟과의 상관관계 (중요!)

target_corr = df_clean[num_cols].corr()['임신 성공 여부'].drop('임신 성공 여부')
target_corr = target_corr.sort_values(ascending=False)

print("\n=== 임신 성공 여부와 상관관계 TOP 10 (양의 상관) ===")
print(target_corr.head(10))

print("\n=== 임신 성공 여부와 상관관계 TOP 10 (음의 상관) ===")
print(target_corr.tail(10))

# 시각화
plt.figure(figsize=(10, 8))
target_corr.sort_values().plot(
    kind='barh',
    color=['#FF6B6B' if x < 0 else '#4ECDC4' for x in target_corr.sort_values()]
)
plt.title('각 변수와 임신 성공 여부 상관관계', fontsize=14)
plt.xlabel('상관계수')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.tight_layout()
plt.savefig('../data/eda_corr_target.png', dpi=100, bbox_inches='tight')
plt.show()


# 3. 전체 상관관계 히트맵

# 상관관계 높은 변수 TOP 15만 추출
top_cols = target_corr.abs().sort_values(ascending=False).head(15).index.tolist()
top_cols.append('임신 성공 여부')

plt.figure(figsize=(14, 12))
sns.heatmap(
    df_clean[top_cols].corr(),
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=0.5
)
plt.title('주요 변수 상관관계 히트맵 (TOP 15)', fontsize=14)
plt.tight_layout()
plt.savefig('../data/eda_heatmap.png', dpi=100, bbox_inches='tight')
plt.show()


# 4. 주요 변수별 성공률 비교

key_cols = target_corr.abs().sort_values(ascending=False).head(5).index.tolist()

fig, axes = plt.subplots(1, 5, figsize=(20, 5))
for i, col in enumerate(key_cols):
    df_clean.groupby('임신 성공 여부')[col].mean().plot(
        kind='bar', ax=axes[i], color=['#FF6B6B', '#4ECDC4']
    )
    axes[i].set_title(f'{col}')
    axes[i].set_xlabel('0:실패 / 1:성공')
    axes[i].tick_params(axis='x', rotation=0)

plt.suptitle('성공/실패 그룹별 주요 변수 평균값 비교', fontsize=14)
plt.tight_layout()
plt.savefig('../data/eda_key_vars.png', dpi=100, bbox_inches='tight')
plt.show()
```

```python
#  ✅ 임신 성공에 긍정적 영향 (양의 상관)
#1위. 이식된 배아 수        +0.159  → 배아 많을수록 성공률 ↑
#2위. 배아 이식 경과일       +0.157  → 이식 타이밍 중요
#3위. 총 생성 배아 수        +0.146  → 배아 많이 만들수록 유리
#4위. 단일 배아 이식 여부     +0.135
#5위. 혼합된 난자 수         +0.116

#❌ 임신 성공에 부정적 영향 (음의 상관)
#1위. 동결 배아 사용 여부   -0.028  → 신선 배아가 더 유리
#2위. 저장된 신선 난자 수   -0.047  → 저장보다 즉시 사용이 유리
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df_clean = pd.read_csv("../data/train_clean.csv")

# 1. 범주형 컬럼 확인

cat_cols = df_clean.select_dtypes(include='object').columns.tolist()
print(f"범주형 컬럼 수: {len(cat_cols)}개")
print(cat_cols)


# 2. Label Encoding

le = LabelEncoder()
df_encoded = df_clean.copy()

for col in cat_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    print(f"[인코딩 완료] {col} → 고유값 수: {df_clean[col].nunique()}개")


# 3. 결과 확인

print("\n=== 인코딩 후 데이터 타입 확인 ===")
print(df_encoded.dtypes.value_counts())

print("\n=== 인코딩 후 상위 3행 ===")
print(df_encoded.head(3))


# 4. 저장

df_encoded.to_csv("../data/train_encoded.csv", index=False, encoding='utf-8-sig')
print("\n train_encoded.csv 저장 완료!")
print(f"최종 데이터 크기: {df_encoded.shape}")
```
